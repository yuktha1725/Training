```python
data=pd.get_dummies(data,columns=['Sex','AgeCategory','Race','GenHealth',],drop_first=True)
print(data.head())
```

```python
X=data.drop(columns=['HeartDisease'])
y=data['HeartDisease'].map({"Yes":1,"No":0})
```
